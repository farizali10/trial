```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.pandas.set_option("display.max_columns",None)
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

class Model_Finder:
    """
    This class is used to find the model with the best R2 score.
    
    Version: 1.0
    Revisions: None
    """

    def __init__(self):
        """
        Constructor method to initialize the model finder.
        """
        self.linearReg = LinearRegression()
        self.RidgeReg = Ridge()
        self.RandomForestReg = RandomForestRegressor()
        self.SVR_model = SVR()
        
    def get_best_params_for_SVR(self, train_x, train_y):
        """
        Method Name: get_best_params_for_SVR
        Description: Get the best parameters for Support Vector Regression (SVR) which give the best R2 score.
                     Use Hyperparameter Tuning with GridSearchCV.
        Output: The model with the best parameters
        On Failure: Raise Exception

        Version: 1.0
        Revisions: None
        """
        try:
            param_grid_svr = {
                "kernel": ['linear', 'poly', 'rbf', 'sigmoid'],
                "degree": [3, 4, 5],
                "gamma": ['scale', 'auto', 0.1, 0.01],
                "coef0": [0.0, 0.1, 0.01],
                "tol": [1e-3, 1e-4],
                "C": [0.1, 1.0, 10.0],
                "epsilon": [0.1, 0.01, 0.001],
                "shrinking": [True, False],
                "cache_size": [100, 200, 300],
                "verbose": [True, False],
                "max_iter": [-1, 1000, 2000]
            }
            grid_search = GridSearchCV(self.SVR, param_grid_svr, verbose=3, cv=5)
            grid_search.fit(train_x, train_y)
            best_svr = grid_search.best_estimator_
            return best_svr
        except Exception as e:
            print('Exception occurred in get_best_params_for_SVR: ' + str(e))
            raise Exception()

    def get_best_params_for_Random_Forest_Regressor(self, train_x, train_y):
        """
        Method Name: get_best_params_for_Random_Forest_Regressor
        Description: Get the best parameters for Random Forest Regressor Algorithm which give the best R2 score.
                     Use Hyperparameter Tuning with GridSearchCV.
        Output: The model with the best parameters
        On Failure: Raise Exception

        Version: 1.0
        Revisions: None
        """
        try:
            param_grid_Random_forest_Tree = {
                "n_estimators": [100, 200, 300],
                "criterion": ["mse", "mae"],
                "max_depth": [None, 5, 10],
                "min_samples_split": [2, 5, 10],
                "min_samples_leaf": [1, 2, 4],
                "max_features": ["auto", "sqrt", "log2"],
                "max_leaf_nodes": [None, 5, 10],
                "min_impurity_decrease": [0.0, 0.1, 0.2],
                "bootstrap": [True, False],
                "oob_score": [True, False],
                "random_state": [None, 42],
                "verbose": [0, 1],
                "warm_start": [True, False],
                "ccp_alpha": [0.0, 0.1, 0.2],
                "max_samples": [None, 0.5, 1.0],
                "monotonic_cst": [None, [1, 0, -1]]
            }
            grid_search = GridSearchCV(self.RandomForestReg, param_grid_Random_forest_Tree, verbose=3, cv=5)
            grid_search.fit(train_x, train_y)
            best_random_forest_reg = grid_search.best_estimator_
            return best_random_forest_reg
        except Exception as e:
            print('Exception occurred in get_best_params_for_Random_Forest_Regressor: ' + str(e))
            raise Exception()
            
    def get_best_params_for_linearReg(self, train_x, train_y):
        """
        Method Name: get_best_params_for_linearReg
        Description: Get the best parameters for Linear Regression Algorithm which give the best R2 score.
                     Use Hyperparameter Tuning with GridSearchCV.
        Output: The model with the best parameters
        On Failure: Raise Exception

        Version: 1.0
        Revisions: None
        """
        try:
            param_grid_linearReg = {
                'fit_intercept': [True, False],
                'copy_X': [True, False],
                'positive': [True, False]
            }
            grid_search = GridSearchCV(self.linearReg, param_grid_linearReg, verbose=3, cv=5)
            grid_search.fit(train_x, train_y)
            best_linear_reg = grid_search.best_estimator_
            return best_linear_reg
        except Exception as e:
            print('Exception occurred in get_best_params_for_linearReg: ' + str(e))
            raise Exception()

    def get_best_params_for_RidgeReg(self, train_x, train_y):
        """
        Method Name: get_best_params_for_RidgeReg
        Description: Get the best parameters for Ridge Regression Algorithm which give the best R2 score.
                     Use Hyperparameter Tuning with GridSearchCV.
        Output: The model with the best parameters
        On Failure: Raise Exception

        Version: 1.0
        Revisions: None
        """
        try:
            param_grid_RidgeReg = {
                'alpha': [0.1, 1, 10, 100],
                'fit_intercept': [True, False],
                'copy_X': [True, False],
                'max_iter': [None, 1000, 15000],
                'tol': [1e-4, 1e-5],
                'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga'],
                'positive': [True, False],
                'random_state': [None, 42]
            }
            grid_search = GridSearchCV(self.RidgeReg, param_grid_RidgeReg, verbose=3, cv=5)
            grid_search.fit(train_x, train_y)
            best_ridge_reg = grid_search.best_estimator_
            return best_ridge_reg
        except Exception as e:
            print('Exception occurred in get_best_params_for_RidgeReg: ' + str(e))
            raise Exception()

    def get_best_model(self, train_x, train_y, test_x, test_y):
        """
        Method Name: get_best_model
        Description: Find the best model among Linear Regression, Ridge Regression, SVR, and RandomForestRegressor
                     based on the R2 score.
        Output: The best model name and the model object
        On Failure: Raise Exception

        Version: 1.0
        Revisions: None
        """
        try:
            # Train models and get predictions
            linear_reg = self.get_best_params_for_linearReg(train_x, train_y)
            ridge_reg = self.get_best_params_for_RidgeReg(train_x, train_y)
            svr_reg = self.get_best_params_for_SVR(train_x, train_y)
            random_forest_reg = self.get_best_params_for_Random_Forest_Regressor(train_x, train_y)

            linear_reg_pred = linear_reg.predict(test_x)
            ridge_reg_pred = ridge_reg.predict(test_x)
            svr_reg_pred = svr_reg.predict(test_x)
            random_forest_reg_pred = random_forest_reg.predict(test_x)

            # Calculate R2 scores
            linear_reg_r2 = r2_score(test_y, linear_reg_pred)
            ridge_reg_r2 = r2_score(test_y, ridge_reg_pred)
            svr_reg_r2 = r2_score(test_y, svr_reg_pred)
            random_forest_reg_r2 = r2_score(test_y, random_forest_reg_pred)

            # Choose the best model
            best_model_name = None
            best_model = None
            best_r2_score = float('-inf')

            if linear_reg_r2 > best_r2_score:
                best_r2_score = linear_reg_r2
                best_model_name = 'Linear Regression'
                best_model = linear_reg
            if ridge_reg_r2 > best_r2_score:
                best_r2_score = ridge_reg_r2
                best_model_name = 'Ridge Regression'
                best_model = ridge_reg
            if svr_reg_r2 > best_r2_score:
                best_r2_score = svr_reg_r2
                best_model_name = 'SVR'
                best_model = svr_reg
            if random_forest_reg_r2 > best_r2_score:
                best_r2_score = random_forest_reg_r2
                best_model_name = 'Random Forest Regressor'
                best_model = random_forest_reg

            return best_model_name, best_model
        except Exception as e:
            print('Exception occurred in get_best_model: ' + str(e))
            raise Exception()
```
